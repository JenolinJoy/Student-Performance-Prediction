# Student Performance Prediction System using Machine Learning

This project predicts whether a student will pass or fail based on academic factors such as attendance, internal marks, and assignment scores.


```python
import pandas as pd
import random

# Generate dataset
data = {
    "Attendance": [],
    "Internal_Marks": [],
    "Assignment_Score": [],
    "Result": []
}

for i in range(60):  # 60 students
    attendance = random.randint(30, 95)
    marks = random.randint(25, 90)
    assignment = random.randint(30, 90)

    if attendance > 50 and marks > 40 and assignment > 45:
        result = "Pass"
    else:
        result = "Fail"

    data["Attendance"].append(attendance)
    data["Internal_Marks"].append(marks)
    data["Assignment_Score"].append(assignment)
    data["Result"].append(result)

df = pd.DataFrame(data)
df.head()
```

```python
df["Result"] = df["Result"].map({"Pass": 1, "Fail": 0})
df.head()
```

```python
X = df[["Attendance", "Internal_Marks", "Assignment_Score"]]
y = df["Result"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Attendance"], df["Result"])
plt.xlabel("Attendance")
plt.ylabel("Result (1 = Pass, 0 = Fail)")
plt.title("Attendance vs Student Performance")
plt.show()
```

## Conclusion

This project demonstrates a machine learning approach to predict student performance using academic parameters. Logistic Regression was used to classify students as pass or fail, achieving good prediction accuracy. The system can help identify students who may require additional academic support.
